```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
topic = "data structures"
url_link = f"https://en.wikipedia.org/wiki/{'_'.join(topic.split(' '))}"
html_doc = urlopen(url_link).read()
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
headers = []
for header in soup.find_all("span", {"class" : "mw-headline"}):
    if header.parent.name != "h2":
        continue
    if header.text[:len("See also")] == "See also":
        break
    headers.append(header.text.replace("[edit]", ""))
print(headers)
```

```python
def clean_sentences(t):
    sentences = t.split(".")
    #print(sentences)
    return [" ".join(s.split()) for s in sentences if len(s) > 10]
```

```python
texts = []

for header in soup.find_all("h2"):

    if header.text.replace("[edit]", "") not in headers:
        continue

    paragraphs = []
    nextNode = header
    while True:
        nextNode = nextNode.nextSibling
        if nextNode.name == "h2":
            break
        if nextNode.name == "p":
            paragraphs.extend(clean_sentences(nextNode.text))
    texts.append(paragraphs)
```

```python
questions = [f"What are {header} of {topic}" for header in headers]
```

```python
data = pd.DataFrame()
data1 = pd.DataFrame({'questions': questions, 'answers': texts})
data = data.append(data1)
data
```

```python
def generate_topic_questions(topic, data):
    
    url_link = f"https://en.wikipedia.org/wiki/{'_'.join(topic.split(' '))}"
    html_doc = urlopen(url_link).read()
    soup = BeautifulSoup(html_doc, 'html.parser')
    
    headers = []
    for header in soup.find_all("span", {"class" : "mw-headline"}):
        if header.parent.name != "h2":
            continue
        if header.text[:len("See also")] == "See also":
            break
        headers.append(header.text.replace("[edit]", ""))
    
    texts = []
    for header in soup.find_all("h2"):

        if header.text.replace("[edit]", "") not in headers:
            continue

        paragraphs = []
        nextNode = header
        while True:
            nextNode = nextNode.nextSibling
            if nextNode.name == "h2":
                break
            if nextNode.name == "p":
                paragraphs.extend(clean_sentences(nextNode.text))
        texts.append(paragraphs)
        
    questions = [f"What are {header} of {topic}" for header in headers]
    
    data1 = pd.DataFrame({'questions': questions, 'answers': texts})
    return data.append(data1)
```

```python
topics = pd.read_csv("Keywords-Springer-83K-20210405.csv", header = None)[0].tolist()[:1000]
```

```python
data = pd.DataFrame()
for topic in topics:
    try:
        data = generate_topic_questions(topic, data)
    except:
        print(f"failed to generate dataset for: {topic}")
```
